```python
import pandas as pd
df = pd.read_csv("../Data/processed/cleaned_ecommerce_customer_churn_data.csv")

print("Dataset loaded successfully")
```

```python
# Feature Engineering
# Tenure Months
df["Tenure_Months"] = df["Membership_Years"] * 12

# Engagement Score Metric
df["Engagement_Score"] = (df["Login_Frequency"] + 
                          df["Session_Duration_Avg"] + 
                          df["Pages_Per_Session"] + 
                          df["Email_Open_Rate"] + 
                          df["Mobile_App_Usage"]
                        )

# Purchase Intensity Total Putchase / Tenure Months
df["Purchase_Intensity"] = df["Total_Purchases"] / (df["Tenure_Months"] + 1)

# Recency Feature
df["Recency_Score"] = 1 / (df["Days_Since_Last_Purchase"] + 1)

# High Risk Customers
df["High_Risk"] = ((df["Cart_Abandonment_Rate"] > 0.5) & (df["Login_Frequency"] < 5)).astype(int)
```
