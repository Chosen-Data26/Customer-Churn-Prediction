```python
# Question: What pattern exists between customer behaviour and churn?
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../Data/processed/cleaned_ecommerce_customer_churn_data.csv")

sns.countplot(x="Churned", data=df)
plt.title("Churn Distribution")
plt.show()
```

The dataset shows a moderate class imbalance, with approximately 71% active users and 29% churned users. While this imbalance is not extreme, it is important to consider during model evaluation to avoid bias toward predicting the majority class.

```python
num_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(12,8))
sns.heatmap(num_df.corr(), cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Churned', y='Login_Frequency', data=df)
plt.show()

sns.boxplot(x='Churned', y='Session_Duration_Avg', data=df)
plt.show()

sns.boxplot(x='Churned', y='Total_Purchases', data=df)
plt.show()

sns.boxplot(x='Churned', y='Days_Since_Last_Purchase', data=df)
plt.show()

sns.boxplot(x='Churned', y='Cart_Abandonment_Rate', data=df)
plt.show()
```

```python
sns.countplot(x='Gender', hue='Churned', data=df)
plt.show()

sns.countplot(x='Country', hue='Churned', data=df)
plt.xticks(rotation=45)
plt.show()
```
